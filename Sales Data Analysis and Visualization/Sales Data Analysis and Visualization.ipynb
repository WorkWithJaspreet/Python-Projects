# Sales Data Analysis and Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('C:/Users/ASUS/superstore_sales.csv')

# Data Preprocessing
data.dropna(inplace=True)  # Remove missing values

# Convert order date to datetime, specifying dayfirst format
data['Order Date'] = pd.to_datetime(data['Order Date'], dayfirst=True)

# Exploratory Data Analysis
total_sales = data['Sales'].sum()
average_sales = data['Sales'].mean()
top_products = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Visualization: Sales Trend Over Time
plt.figure(figsize=(10, 6))
data.groupby('Order Date')['Sales'].sum().plot(kind='line')
plt.title('Sales Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

# Visualization: Total Sales by Product Category
plt.figure(figsize=(10, 6))
data.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Total Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

# Visualization: Sales Distribution by Region
plt.figure(figsize=(10, 6))
data.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Region')
plt.ylabel('')
plt.show()

# Additional Analysis: Top 10 Products by Sales
plt.figure(figsize=(12, 8))
top_products.plot(kind='bar')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
